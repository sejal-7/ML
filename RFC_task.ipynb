```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\lenovo\Desktop\Python programs\Data\creditcard.csv")
df.head()
```

```python
X = df.drop("Class", axis = "columns")
X.head()
```

```python
y = df.Class
y.head()
```

```python
fraud = y[y == 1].value_counts()[1]
print(fraud)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
m1 = RandomForestClassifier(n_estimators = 5, criterion = 'entropy')
```

```python
m1.fit(X_train, y_train)
```

```python
m1.score(X_test, y_test)
```

```python
a= m1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test,a)
```
